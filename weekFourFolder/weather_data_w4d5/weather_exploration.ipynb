```python
def clean_data(original):
    clean_df = original[original['STATION'] == 'USW00094728'].copy()
    clean_df = clean_df.dropna(axis=1, how='any')
    clean_df['DATE'] = pd.to_datetime(clean_df['DATE'])
    return clean_df
```

```python
original_df2 = pd.read_csv("1-1-2020_12-31-2022.csv")
original_df3 = pd.read_csv("1-1-2023_today.csv")
```

```python
df2 = clean_data(original_df2)
df3 = clean_data(original_df3)
```

```python
df2.info()
```

```python
df3.info()
```

```python
df = pd.concat([df1, df2, df3], ignore_index=True)
df.info()
```

```python

df = pd.read_csv("6-15-2017_12-31-2019.csv")  # Replace with your actual filename
rainy_days = df[df['PRCP'] > 0].copy()
print(df['PRCP'].head())
print(df['PRCP'].dtype)
print(df['PRCP'].isnull().sum())
```

```python
df['PRCP'] = pd.to_numeric(df['PRCP'], errors='coerce')
df['PRCP'] = df['PRCP'].fillna(0)
print(rainy_days.head())
print(f"There are {len(rainy_days)} rainy days in this dataset.")
```

```python
max_rain = rainy_days[rainy_days['PRCP'] == rainy_days['PRCP'].max()]
print(max_rain[['DATE', 'PRCP']])
avg_rain = rainy_days['PRCP'].mean()
print(f"Average rainfall on rainy days: {avg_rain:.2f} inches")

```

```python
import matplotlib.pyplot as plt

rainy_days['DATE'] = pd.to_datetime(rainy_days['DATE'])
plt.figure(figsize=(12, 4))
plt.plot(rainy_days['DATE'], rainy_days['PRCP'], alpha=0.5)
plt.title('Rainfall Over Time')
plt.xlabel('Date')
plt.ylabel('Precipitation (inches)')
plt.tight_layout()
plt.show()
```

```python
# Make sure DATE is in datetime format
rainy_days['DATE'] = pd.to_datetime(rainy_days['DATE'])

# Create the scatter plot
plt.figure(figsize=(12, 4))
plt.scatter(rainy_days['DATE'], rainy_days['PRCP'], alpha=0.5, color='blue', s=10)

# Add labels and title
plt.title('Rainfall Over Time (Scatter Plot)')
plt.xlabel('Date')
plt.ylabel('Precipitation (inches)')
plt.tight_layout()
plt.show()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')

df["YEAR_MONTH"] = df["DATE"].dt.to_period('M')
df.info()
```

```python
df.head()
```

```python
monthly_rain_df = df.groupby('YEAR_MONTH')['PRCP'].sum().reset_index()
monthly_rain_df['YEAR_MONTH'] = monthly_rain_df['YEAR_MONTH'].dt.to_timestamp()
monthly_rain_df.info()
```

```python
monthly_rain_df.head()
```

```python
plt.figure(figsize=(12, 6))

# If YEAR_MONTH is a Period type, convert to string for better plotting
x = monthly_rain_df['YEAR_MONTH'].astype(str)
y = monthly_rain_df['PRCP']

plt.scatter(x, y, alpha=0.6, color='blue')
plt.title('Monthly Rainfall Totals')
plt.xlabel('Year-Month')
plt.ylabel('Total Precipitation (inches)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
