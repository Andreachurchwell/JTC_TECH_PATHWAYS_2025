```python
# Step 1: Filter for medal winners
medal_winners = df[df['Medal'].notna()]

# Step 2: Find max height among medal winners
max_height = medal_winners['Height'].max()

# Step 3: Return all tallest medal winners
tallest_medalists = medal_winners[medal_winners['Height'] == max_height][['Name', 'Height', 'Medal']].drop_duplicates()

print(tallest_medalists)
```

```python
tallest_medalists = tallest_medalists[tallest_medalists['Medal'] == 'Gold']
print(tallest_medalists)
```

```python
medal_winners = df[df['Medal'].notna()]

# Step 2: Find max height among medal winners
min_height = medal_winners['Height'].min()

# Step 3: Return all tallest medal winners
shortest_medalists = medal_winners[medal_winners['Height'] == min_height][['Name', 'Height', 'Medal']].drop_duplicates("Name")

print(shortest_medalists)


```
